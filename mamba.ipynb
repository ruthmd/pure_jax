```python
import jax
import jax.numpy as jnp
from jax import random, jit, vmap, grad
import flax.linen as nn
import optax
from flax.training import train_state
import numpy as np
from typing import Tuple, Dict, Any, Optional, List, Callable
import math
import functools
import time
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
from datasets import load_dataset
from transformers import AutoTokenizer
import pickle
import os

# Enable mixed precision for better memory efficiency and speed
# Comment this out if causing issues on your GPU
os.environ['XLA_FLAGS'] = '--xla_gpu_enable_fast_min_max'
try:
    jax.config.update('jax_default_dtype_bits', 16)  # Use BF16
    print("Mixed precision enabled")
except:
    print("Mixed precision not enabled - continuing with full precision")

def setup_jax_devices():
    """
    Configure JAX to use GPU if available, otherwise use CPU.
    """
    # Get available devices
    devices = jax.devices()
    
    # Check for GPU
    gpu_devices = [d for d in devices if d.platform == 'gpu']
    tpu_devices = [d for d in devices if d.platform == 'tpu']
    
    device_info = {
        'total_devices': len(devices),
        'has_gpu': bool(gpu_devices),
        'has_tpu': bool(tpu_devices),
        'active_device': None,
    }
    
    # Set XLA flags to enable memory growth (prevents OOM on GPU)
    os.environ['XLA_PYTHON_CLIENT_PREALLOCATE'] = 'false'
    os.environ['XLA_PYTHON_CLIENT_ALLOCATOR'] = 'platform'
    
    # Print device information
    if device_info['has_gpu']:
        print(f"GPU detected: {len(gpu_devices)} device(s)")
        print(f"Device: {gpu_devices[0]}")
        device_info['active_device'] = 'gpu'
    elif device_info['has_tpu']:
        print(f"TPU detected: {len(tpu_devices)} device(s)")
        print(f"Device: {tpu_devices[0]}")
        device_info['active_device'] = 'tpu'
    else:
        print("No GPU or TPU detected, using CPU")
        print(f"Device: {devices[0]}")
        device_info['active_device'] = 'cpu'
    
    return device_info



def softplus_with_beta(x, beta=1.0):
    """
    Custom implementation of softplus with beta parameter.
    
    Args:
        x: Input tensor
        beta: Beta parameter for scaling the input
        
    Returns:
        softplus(beta * x) / beta
    """
    # For numerical stability, avoid overflow
    scaled_x = x * beta
    # Use a safe implementation to avoid overflow/underflow
    is_large = scaled_x > 15.0  # For large x, exp(x) is close to infinity
    safe_exp = jnp.where(is_large, scaled_x, jnp.log(1.0 + jnp.exp(scaled_x)))
    result = jnp.where(is_large, safe_exp, safe_exp) / beta
    return result


class SelectiveScanModule(nn.Module):
    """
    Core selective scan mechanism of Mamba.
    
    This implements the selective SSM mechanism with improved initialization
    and numerical stability enhancements.
    """
    d_model: int
    d_state: int
    dt_rank: Optional[int] = None
    dt_min: float = 0.001
    dt_max: float = 0.1
    dt_init: str = "random"
    dt_scale: float = 1.0
    dt_init_floor: float = 1e-4
    
    def setup(self):
        # Calculate dt_rank without modifying the attribute
        dt_rank = self.dt_rank if self.dt_rank is not None else math.ceil(self.d_model / 16)
        
        # Improved initialization for better training stability
        dt_init = jnp.linspace(
            jnp.log(self.dt_min),
            jnp.log(self.dt_max),
            dt_rank
        )
        
        # Parameters with better initialization
        self.dt_proj = self.param('dt_proj', 
                                  lambda rng, shape: jax.random.normal(rng, shape) * 0.01,  # Reduced variance
                                  (self.d_model, dt_rank))
        self.dt_param = self.param('dt_param', lambda _: dt_init)
        
        # A, B, C, D parameters - use improved initialization
        self.A_log = self.param('A_log', 
                                lambda rng, shape: jnp.ones(shape) * -0.5,  # More stable range
                                (self.d_state,))
        self.D = self.param('D', 
                            lambda rng, shape: jnp.ones(shape) * 0.5,  # Scale down initial skip connection
                            (self.d_model,))
        
        # State projections with improved initialization
        self.B_proj = nn.Dense(
            self.d_state,
            kernel_init=nn.initializers.variance_scaling(
                scale=1.0, mode='fan_in', distribution='truncated_normal'
            )
        )
        
        self.C_proj = nn.Dense(
            self.d_model,
            kernel_init=nn.initializers.variance_scaling(
                scale=1.0, mode='fan_out', distribution='truncated_normal'
            )
        )
        
    def __call__(self, u, delta, state=None, return_state=False):
        # Get shapes
        batch, seq_len, _ = u.shape
        _, delta_seq_len, _ = delta.shape
        
        # Check if sequence lengths match
        if seq_len != delta_seq_len:
            # Resize delta to match u's sequence length
            if delta_seq_len > seq_len:
                # Truncate
                delta = delta[:, :seq_len, :]
            else:
                # Pad with zeros
                padding = jnp.zeros((batch, seq_len - delta_seq_len, delta.shape[-1]))
                delta = jnp.concatenate([delta, padding], axis=1)
        
        if state is None:
            # Initialize state to zeros
            state = jnp.zeros((batch, self.d_state))
        
        # Project u to state dimensions
        B = self.B_proj(u)  # (B, L, d_state)
        
        # Calculate dt with improved numerical stability
        x = jnp.einsum('bld,de->ble', delta, self.dt_proj)
        x = x + self.dt_param
        
        # Use standard softplus without beta parameter
        # dt = self.dt_scale * jax.nn.softplus(x)
        
        # Use our custom softplus with beta implementation
        dt = self.dt_scale * softplus_with_beta(x, beta=1.0)
        
        # Average across dt dimensions to get a scalar per position
        dt = jnp.mean(dt, axis=-1, keepdims=True)  # (B, L, 1)
        
        # Compute dA with improved stability
        A_log_expanded = jnp.reshape(self.A_log, (1, 1, -1))  # (1, 1, d_state)
        # Use clipping to prevent extreme values
        A_log_expanded = jnp.clip(A_log_expanded, -10.0, 0.0)
        dA = jnp.exp(A_log_expanded * dt)  # (B, L, d_state)
        
        # Compute dB
        dB = B * dt  # (B, L, d_state)
        
        # Prepare for scan - reshape tensors for scan's expected format
        def scan_fn(h, inputs):
            dA_t, dB_t = inputs
            h_next = dA_t * h + dB_t
            return h_next, h_next
        
        # Transpose tensors for scan (sequence dim first)
        dA_time = jnp.transpose(dA, (1, 0, 2))  # (L, B, d_state)
        dB_time = jnp.transpose(dB, (1, 0, 2))  # (L, B, d_state)
        
        # Run scan without precision parameter
        _, states = jax.lax.scan(scan_fn, state, (dA_time, dB_time))
        
        # Transpose states back to batch-first
        states = jnp.transpose(states, (1, 0, 2))  # (B, L, d_state)
        
        # Project to output space
        y = self.C_proj(states)  # (B, L, d_model)
        
        # Add residual connection with scaling
        y = y + u * self.D
        
        if return_state:
            return y, states[:, -1, :]
        return y


class MambaBlock(nn.Module):
    """
    Mamba block with improved architecture:
    1. Layer normalization with better stability
    2. Optimized convolution settings
    3. Enhanced gating mechanism
    4. Improved residual connections
    """
    d_model: int
    d_state: int = 16
    d_conv: int = 4
    expand: int = 1  # Reduced expansion factor
    dt_rank: Optional[int] = None
    conv_kernel: int = 4
    activation: str = "silu"
    dropout_rate: float = 0.0
    layer_idx: int = 0  # For potential layer-specific adjustments
    
    def setup(self):
        self.d_inner = self.expand * self.d_model
        
        # Pre-normalization - use RMSNorm for better stability
        self.norm = nn.LayerNorm(epsilon=1e-5)  # Can alternatively use RMSNorm if preferred
        
        # Input projection
        # Use custom scale initialization 
        scale = 0.02 * (1.0 / max(1.0, math.sqrt(self.layer_idx + 1.0)))
        self.in_proj = nn.Dense(
            self.d_inner * 2,  # For gating
            kernel_init=nn.initializers.normal(stddev=scale),
            bias_init=nn.initializers.zeros
        )
        
        # Convolution with better padding
        padding = self.conv_kernel - 1
        self.conv = nn.Conv(
            features=self.d_inner,
            kernel_size=(self.conv_kernel,),
            padding=((padding, 0),),
            feature_group_count=1,
            use_bias=True,  # Enable bias for better expressivity
            kernel_init=nn.initializers.variance_scaling(
                scale=2.0, mode='fan_in', distribution='truncated_normal'
            )
        )
        
        # SSM with improved state dimension
        self.ssm = SelectiveScanModule(
            d_model=self.d_inner,
            d_state=self.d_state,
            dt_rank=self.dt_rank
        )
        
        # Output projection with scaled initialization
        self.out_proj = nn.Dense(
            self.d_model,
            kernel_init=nn.initializers.variance_scaling(
                scale=1.0/math.sqrt(self.layer_idx + 1.0), 
                mode='fan_out', 
                distribution='truncated_normal'
            ),
            bias_init=nn.initializers.zeros
        )
        
        # Conditional dropout
        if self.dropout_rate > 0.0:
            self.dropout = nn.Dropout(rate=self.dropout_rate)
        
        # Activation
        if self.activation == "silu":
            self.act = nn.silu
        elif self.activation == "gelu":
            self.act = nn.gelu
        elif self.activation == "relu":
            self.act = nn.relu
        else:
            raise ValueError(f"Unsupported activation: {self.activation}")
        
    def __call__(self, x, deterministic=True, state=None, return_state=False):
        """
        Args:
            x: input tensor of shape (B, L, D)
            deterministic: whether to use deterministic behavior
            state: optional recurrent state
            return_state: whether to return final state
        """
        skip = x
        x = self.norm(x)
        
        # Input projection and split for SSM and gating
        x_proj_gate = self.in_proj(x)
        x_proj, x_gate = jnp.split(x_proj_gate, 2, axis=-1)
        
        # Apply convolution for local context
        # Transpose for conv operation (batch, channels, length)
        x_conv_input = jnp.transpose(x_proj, (0, 2, 1))
        x_conv_output = self.conv(x_conv_input)
        x_conv = jnp.transpose(x_conv_output, (0, 2, 1))
        
        # Fix dimensions if needed (ensure same sequence length)
        if x_conv.shape[1] != x_gate.shape[1]:
            # Adjust x_conv to match x_gate's sequence length
            seq_len = min(x_conv.shape[1], x_gate.shape[1])
            x_conv = x_conv[:, :seq_len, :]
            x_gate = x_gate[:, :seq_len, :]
        
        # Run SSM
        if return_state:
            ssm_out, new_state = self.ssm(x_conv, x_gate, state=state, return_state=True)
            x_ssm = self.act(ssm_out)
        else:
            x_ssm = self.act(self.ssm(x_conv, x_gate))
        
        # Improved gating mechanism - use sigmoid instead of silu for gate
        gate_factor = jax.nn.sigmoid(x_gate)  # More stable than silu for gating
        gated_output = x_ssm * gate_factor
        
        # Output projection
        output = self.out_proj(gated_output)
        
        # Apply dropout if training
        if self.dropout_rate > 0.0 and not deterministic:
            output = self.dropout(output, deterministic=deterministic)
        
        # Adjust output size if needed to match skip connection
        if output.shape[1] != skip.shape[1]:
            # Adjust output to match skip's sequence length
            seq_len = min(output.shape[1], skip.shape[1])
            output = output[:, :seq_len, :]
            skip = skip[:, :seq_len, :]
        
        # Residual connection with scaling factor for better gradient flow
        # Use a smaller scalar for deeper layers to prevent exploding gradients
        output = output * (0.9 if self.layer_idx >= 1 else 1.0) + skip
        
        if return_state:
            return output, new_state
        return output


class MambaSequenceEncoder(nn.Module):
    """
    Mamba model for encoding text sequences with improved architecture.
    """
    vocab_size: int
    d_model: int = 96  # Reduced from 128
    n_layer: int = 3   # Reduced from 4 
    d_state: int = 12  # Reduced from 16
    expand: int = 1    # Reduced from 2
    use_pretrained: bool = False
    dropout_rate: float = 0.1
    embedding_dropout: float = 0.1
    
    def setup(self):
        # Token embedding with better initialization
        if self.use_pretrained:
            # Load pre-trained embedding weights (this is pseudo-code)
            pretrained_weights = jnp.ones((self.vocab_size, self.d_model))
            self.embedding = nn.Embed(
                self.vocab_size, 
                self.d_model,
                embedding_init=lambda _, __: pretrained_weights
            )
        else:
            # Use scaled normal initialization for embeddings
            self.embedding = nn.Embed(
                self.vocab_size, 
                self.d_model,
                embedding_init=nn.initializers.normal(stddev=0.02)
            )
        
        # Add embedding dropout
        self.emb_dropout = nn.Dropout(rate=self.embedding_dropout)
        
        # Mamba blocks with layer-specific configuration
        self.layers = [
            MambaBlock(
                d_model=self.d_model,
                d_state=self.d_state,
                expand=self.expand,
                dropout_rate=self.dropout_rate * (i + 1) / self.n_layer,  # Progressive dropout
                layer_idx=i  # For layer-specific configurations
            )
            for i in range(self.n_layer)
        ]
        
        # Final layer norm
        self.norm_f = nn.LayerNorm(epsilon=1e-5)  # Increased epsilon for stability
        
    def __call__(self, input_ids, attention_mask=None, deterministic=True):
        """
        Encode input sequence.
        
        Args:
            input_ids: token ids of shape (B, L)
            attention_mask: mask for padding (optional)
            deterministic: whether to use deterministic behavior
            
        Returns:
            x: encoded sequence of shape (B, L, d_model)
        """
        # Get token embeddings
        x = self.embedding(input_ids)
        
        # Apply embedding dropout during training
        if not deterministic:
            x = self.emb_dropout(x, deterministic=deterministic)
        
        # Apply mask if provided
        if attention_mask is not None:
            x = x * attention_mask[:, :, None]
            
        # Apply Mamba blocks
        for layer in self.layers:
            x = layer(x, deterministic=deterministic)
                
        # Final normalization
        x = self.norm_f(x)
        
        # Apply mask again to ensure padded tokens don't contribute
        if attention_mask is not None:
            x = x * attention_mask[:, :, None]
        
        return x


class MambaForSequenceClassification(nn.Module):
    """
    Mamba model for sequence classification (emotion detection) with improved architecture.
    """
    vocab_size: int  # Size of vocabulary
    num_labels: int  # Number of classes
    d_model: int = 96  # Model dimension (reduced from 128)
    n_layer: int = 3  # Number of layers (reduced from 4)
    d_state: int = 12  # State dimension (reduced from 16)
    expand: int = 1  # Expansion factor (reduced from 2)
    dropout_rate: float = 0.3  # Increased dropout rate
    classifier_dropout: float = 0.5  # Separate dropout for classifier head
    embedding_dropout: float = 0.1  # Dropout for embeddings
    label_smoothing: float = 0.1  # Label smoothing factor
    
    def setup(self):
        # Sequence encoder
        self.encoder = MambaSequenceEncoder(
            vocab_size=self.vocab_size,
            d_model=self.d_model,
            n_layer=self.n_layer,
            d_state=self.d_state,
            expand=self.expand,
            dropout_rate=self.dropout_rate,
            embedding_dropout=self.embedding_dropout
        )
        
        # Classification head with improved structure
        self.pre_classifier = nn.Dense(
            self.d_model,  # Create bottleneck layer 
            kernel_init=nn.initializers.variance_scaling(
                scale=2.0, mode='fan_in', distribution='truncated_normal'
            )
        )
        self.activation = nn.gelu  # Use GELU instead of ReLU
        self.dropout = nn.Dropout(self.classifier_dropout)
        self.classifier = nn.Dense(
            self.num_labels,
            kernel_init=nn.initializers.variance_scaling(
                scale=1.0, mode='fan_out', distribution='truncated_normal'
            )
        )
        self.layer_norm = nn.LayerNorm(epsilon=1e-5)  # Add layer norm before classifier
        
    def __call__(self, input_ids, attention_mask=None, deterministic=True):
        """
        Run model for classification.
        
        Args:
            input_ids: token ids of shape (B, L)
            attention_mask: mask for padding (optional)
            deterministic: whether to use deterministic behavior
            
        Returns:
            logits: classification logits of shape (B, num_labels)
        """
        # Encode sequence
        sequence_output = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            deterministic=deterministic
        )
        
        # Improved pooling strategy - use attention-weighted mean pooling
        if attention_mask is not None:
            # Create attention weights for weighted pooling
            weights = sequence_output @ jnp.ones((self.d_model, 1))  # (B, L, 1)
            weights = jnp.where(
                attention_mask[:, :, None] > 0,
                jax.nn.softmax(weights, axis=1),
                0.0
            )
            
            # Apply weighted pooling
            pooled_output = jnp.sum(weights * sequence_output, axis=1)  # (B, d_model)
        else:
            # Simple mean pooling if no mask
            pooled_output = jnp.mean(sequence_output, axis=1)
        
        # Apply pre-classifier transformation
        pooled_output = self.pre_classifier(pooled_output)
        pooled_output = self.activation(pooled_output)
        pooled_output = self.layer_norm(pooled_output)  # Add layer norm for stability
        
        # Apply dropout during training
        if not deterministic:
            pooled_output = self.dropout(pooled_output, deterministic=deterministic)
        
        # Final classification layer
        logits = self.classifier(pooled_output)
        
        return logits


#----------------------------------------
# Training Utilities with Improvements
#----------------------------------------

class TrainState(train_state.TrainState):
    """
    Extended TrainState with dropout RNG and EMA parameters.
    """
    dropout_rng: jnp.ndarray
    ema_params: Optional[Dict] = None
    ema_decay: float = 0.999


def create_train_state(
    model: nn.Module,
    rng: jnp.ndarray,
    learning_rate: float,
    weight_decay: float,
    input_shape: Tuple[int, ...],
    warmup_steps: int = 100,
    decay_steps: int = 10000,
    label_smoothing: float = 0.1,
    use_ema: bool = True
) -> TrainState:
    """Creates initial training state with improved optimizer and learning rate schedule."""
    # Split RNG
    rng, dropout_rng = jax.random.split(rng)
    rng, init_rng = jax.random.split(rng)
    
    # Create dummy input
    dummy_input = jnp.ones(input_shape, dtype=jnp.int32)
    dummy_mask = jnp.ones((input_shape[0], input_shape[1]), dtype=jnp.float32)
    
    # Initialize params - don't pass label_smoothing here
    variables = model.init(
        {'params': init_rng, 'dropout': dropout_rng},
        dummy_input,
        dummy_mask,
        deterministic=True  # Use deterministic mode for initialization
    )
    params = variables['params']
    
    # Create improved learning rate schedule with warmup and cosine decay
    if warmup_steps > 0 and decay_steps > 0:
        # Start with a lower learning rate
        init_value = learning_rate / 100.0
        
        # Create a warmup schedule with linear warmup
        warmup_fn = optax.linear_schedule(
            init_value=init_value,
            end_value=learning_rate,
            transition_steps=warmup_steps
        )
        
        # Create a decay schedule with cosine decay and minimum value
        decay_fn = optax.cosine_decay_schedule(
            init_value=learning_rate,
            decay_steps=decay_steps,
            alpha=0.1  # End value is 10% of peak
        )
        
        # Join schedules
        schedule_fn = optax.join_schedules(
            schedules=[warmup_fn, decay_fn],
            boundaries=[warmup_steps]
        )
    else:
        schedule_fn = learning_rate
    
    # Create improved optimizer with gradient clipping and L2 regularization
    # Use Adam instead of Adafactor since your version doesn't support weight_decay
    tx = optax.chain(
        optax.clip_by_global_norm(1.0),  # Gradient clipping
        # Apply L2 regularization separately since Adafactor doesn't support weight_decay
        optax.add_decayed_weights(weight_decay),
        optax.adam(
            learning_rate=schedule_fn,
            b1=0.9,
            b2=0.999,
            eps=1e-8
        )
    )
    
    # Initialize EMA parameters if requested
    ema_params = params if use_ema else None
    
    # Create train state
    return TrainState.create(
        apply_fn=model.apply,
        params=params,
        tx=tx,
        dropout_rng=dropout_rng,
        ema_params=ema_params,
        ema_decay=0.999
    )


@jax.jit
def compute_metrics(logits, labels, label_smoothing=0.1):
    """
    Compute metrics for classification with label smoothing.
    
    Args:
        logits: Model logits
        labels: Ground truth labels
        label_smoothing: Label smoothing factor
        
    Returns:
        metrics: Dictionary of metrics
    """
    # One-hot encode labels
    num_classes = logits.shape[-1]
    one_hot_labels = jax.nn.one_hot(labels, num_classes)
    
    # Apply label smoothing
    smooth_targets = (1.0 - label_smoothing) * one_hot_labels + \
                      label_smoothing / num_classes
    
    # Compute loss with label smoothing
    loss = optax.softmax_cross_entropy(logits, smooth_targets).mean()
    
    # Compute accuracy
    preds = jnp.argmax(logits, axis=-1)
    accuracy = jnp.mean(preds == labels)
    
    return {
        'loss': loss,
        'accuracy': accuracy,
        'predictions': preds
    }


@jax.jit
def update_ema(state: TrainState) -> TrainState:
    """Update exponential moving average of parameters."""
    # Skip if not using EMA
    if state.ema_params is None:
        return state
    
    # Update EMA parameters
    new_ema_params = jax.tree_util.tree_map(
        lambda ema, p: ema * state.ema_decay + p * (1 - state.ema_decay),
        state.ema_params, state.params
    )
    
    # Return updated state
    return state.replace(ema_params=new_ema_params)


@jax.jit
def train_step(
    state: TrainState,
    batch: Dict,
    label_smoothing: float = 0.1,
) -> Tuple[TrainState, Dict]:
    """
    Performs a single training step with improved stability.
    
    Args:
        state: Current train state
        batch: Batch of data
        label_smoothing: Label smoothing factor
        
    Returns:
        state: Updated train state
        metrics: Training metrics
    """
    input_ids = batch['input_ids']
    attention_mask = batch.get('attention_mask', None)
    labels = batch['labels']
    
    dropout_rng = state.dropout_rng
    
    # Define loss function
    def loss_fn(params):
        logits = state.apply_fn(
            {'params': params},
            input_ids,
            attention_mask,
            deterministic=False,
            rngs={'dropout': dropout_rng}
        )
        
        # Compute metrics with label_smoothing as a separate argument
        metrics = compute_metrics(logits, labels, label_smoothing)
        
        return metrics['loss'], metrics
    
    # Compute gradients with improved precision
    grad_fn = jax.value_and_grad(loss_fn, has_aux=True)
    (_, metrics), grads = grad_fn(state.params)
    
    # Update parameters
    new_state = state.apply_gradients(grads=grads)
    
    # Update dropout RNG
    new_dropout_rng = jax.random.fold_in(dropout_rng, 1)
    new_state = new_state.replace(dropout_rng=new_dropout_rng)
    
    # Update EMA parameters
    new_state = update_ema(new_state)
    
    return new_state, metrics


@jax.jit
def eval_step(
    state: TrainState,
    batch: Dict,
    use_ema: bool = True
) -> Dict:
    """
    Performs a single evaluation step.
    
    Args:
        state: Current train state
        batch: Batch of data
        use_ema: Whether to use EMA parameters
        
    Returns:
        metrics: Evaluation metrics
    """
    input_ids = batch['input_ids']
    attention_mask = batch.get('attention_mask', None)
    labels = batch['labels']
    
    # When using JIT, we can't have dynamic conditionals on use_ema
    # Instead, we'll use a static approach with tree_map to select parameters
    
    # Create a mask where True values come from ema_params and False from regular params
    # This mask doesn't depend on evaluating the condition at runtime
    ema_is_present = state.ema_params is not None
    
    # Fixed version that avoids a dynamic conditional
    # This uses the fact that jnp.where evaluates both branches, so it's JIT-compatible
    def select_with_static_mask(regular_param, ema_param):
        # Return ema_param if both use_ema and ema_is_present are True, otherwise regular_param
        return jnp.where(use_ema & ema_is_present, ema_param, regular_param)
    
    # If ema_params exists, apply the selection function to choose between them
    if ema_is_present:
        params = jax.tree_util.tree_map(
            select_with_static_mask, 
            state.params, 
            state.ema_params
        )
    else:
        # If no ema_params, just use regular params
        params = state.params
    
    # Forward pass (deterministic)
    logits = state.apply_fn(
        {'params': params},
        input_ids,
        attention_mask,
        deterministic=True,
        rngs={'dropout': state.dropout_rng}
    )
    
    # Compute metrics with label_smoothing=0.0 for evaluation
    metrics = compute_metrics(logits, labels, label_smoothing=0.0)
    
    return metrics



# def eval_step_no_jit(
#     state: TrainState,
#     batch: Dict,
#     use_ema: bool = True
# ) -> Dict:
#     """
#     Non-JIT version of the evaluation step.
    
#     Args:
#         state: Current train state
#         batch: Batch of data
#         use_ema: Whether to use EMA parameters
        
#     Returns:
#         metrics: Evaluation metrics
#     """
#     input_ids = batch['input_ids']
#     attention_mask = batch.get('attention_mask', None)
#     labels = batch['labels']
    
#     # Simple conditional is fine without JIT
#     params = state.ema_params if use_ema and state.ema_params is not None else state.params
    
#     # Forward pass (deterministic)
#     logits = state.apply_fn(
#         {'params': params},
#         input_ids,
#         attention_mask,
#         deterministic=True,
#         rngs={'dropout': state.dropout_rng}
#     )
    
#     # Compute metrics with label_smoothing=0.0 for evaluation
#     metrics = compute_metrics(logits, labels, label_smoothing=0.0)
    
#     return metrics


#----------------------------------------
# Dataset Loading and Processing with Improvements
#----------------------------------------

def load_emotion_dataset():
    """
    Load the emotion dataset from Hugging Face.
    
    Returns:
        dataset: Dictionary with train, validation, and test splits
    """
    # Load dataset from Hugging Face
    dataset = load_dataset("dair-ai/emotion")
    
    # Return dataset splits
    return {
        'train': dataset['train'],
        'validation': dataset['validation'],
        'test': dataset['test']
    }


def simple_text_augmentation(text, label, seed=None):
    """
    Perform simple text augmentation for better generalization.
    
    Args:
        text: Input text
        label: Label for the text
        seed: Random seed for reproducibility
        
    Returns:
        augmented_text: Augmented text
    """
    if seed is not None:
        np.random.seed(seed)
    
    # Only augment with some probability
    if np.random.random() > 0.5:
        return text
    
    words = text.split()
    
    # Skip if too short
    if len(words) < 4:
        return text
    
    # Choose augmentation type
    aug_type = np.random.choice(['swap', 'delete', 'emoji'])
    
    if aug_type == 'swap' and len(words) >= 4:
        # Swap two random words
        idx1, idx2 = np.random.choice(len(words), 2, replace=False)
        words[idx1], words[idx2] = words[idx2], words[idx1]
    
    elif aug_type == 'delete' and len(words) >= 5:
        # Delete a random word
        idx = np.random.randint(0, len(words))
        words.pop(idx)
    
    elif aug_type == 'emoji':
        # Add emotion-appropriate emoji
        emojis = {
            0: [" 😢", " 😭", " 😔"],  # sadness
            1: [" 😊", " 😁", " 🎉"],  # joy
            2: [" ❤️", " 😍", " 🥰"],  # love
            3: [" 😡", " 😠", " 😤"],  # anger
            4: [" 😨", " 😱", " 😰"],  # fear
            5: [" 😲", " 😯", " 😮"]   # surprise
        }
        
        # Only add emoji if it's an appropriate emotion
        if label in emojis:
            emoji = np.random.choice(emojis[label])
            words.append(emoji)
    
    return " ".join(words)


def prepare_dataset_for_training(
    dataset,
    tokenizer,
    max_length=96,  # Reduced from 128
    batch_size=32,
    shuffle=True,
    augment=True,
    seed=42
):
    """
    Prepare dataset for training with improved preprocessing.
    
    Args:
        dataset: Dataset from Hugging Face
        tokenizer: Tokenizer for encoding text
        max_length: Maximum sequence length
        batch_size: Batch size
        shuffle: Whether to shuffle the dataset
        augment: Whether to apply data augmentation
        seed: Random seed
        
    Returns:
        dataloaders: Dictionary with train, validation, and test dataloaders
        num_batches: Dictionary with number of batches per dataloader
    """
    # Set random seed
    np.random.seed(seed)
    
    # Tokenize function with better handling of special tokens and texts
    def tokenize_function(examples):
        # Clean and standardize text
        texts = [
            text.strip().replace('\n', ' ').replace('\r', ' ')
            for text in examples['text']
        ]
        
        # Apply augmentation only to training data if enabled
        if augment and 'split' in examples and examples.get('split', [''])[0] == 'train':
            # Use the seed to make augmentations reproducible but varied
            seeds = np.arange(len(texts)) + seed
            texts = [
                simple_text_augmentation(text, label, seed=s)
                for text, label, s in zip(texts, examples['label'], seeds)
            ]
        
        # Tokenize with truncation and padding - use dynamic padding for efficiency
        tokenized = tokenizer(
            texts,
            max_length=max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_token_type_ids=False  # No need for token type IDs
        )
        
        return {
            'input_ids': tokenized['input_ids'],
            'attention_mask': tokenized['attention_mask'],
            'labels': examples['label']
        }
    
    # Add split information for selective augmentation
    def add_split_info(examples, split_name):
        examples['split'] = [split_name] * len(examples['text'])
        return examples
    
    # Process each split with proper tracking for augmentation
    tokenized_datasets = {}
    for split in dataset:
        # Add split information
        tagged_dataset = dataset[split].map(
            lambda examples: add_split_info(examples, split),
            batched=True
        )
        
        # Tokenize
        tokenized = tagged_dataset.map(
            tokenize_function,
            batched=True,
            remove_columns=['text', 'split']
        )
        
        # Convert to numpy arrays
        tokenized = tokenized.with_format("numpy")
        tokenized_datasets[split] = tokenized
    
    # Create dataloaders with improved batching
    def create_dataloader(tokenized_dataset, is_train=False):
        # Get data
        input_ids = np.array(tokenized_dataset['input_ids'])
        attention_mask = np.array(tokenized_dataset['attention_mask'])
        labels = np.array(tokenized_dataset['labels'])
        
        # Dataset size
        dataset_size = len(input_ids)
        
        # Calculate class weights for balanced sampling (only for training)
        if is_train:
            class_counts = np.bincount(labels)
            class_weights = 1.0 / (class_counts + 1e-6)  # Add small epsilon to avoid division by zero
            sample_weights = class_weights[labels]
            
            # Normalize weights
            sample_weights = sample_weights / np.sum(sample_weights) * len(sample_weights)
        
        # Create dataloader function
        def dataloader():
            # Shuffle indices with weighted sampling for training
            indices = np.arange(dataset_size)
            if is_train and shuffle:
                np.random.seed(seed)
                if hasattr(np.random, 'choice'):
                    # Use weighted sampling for training to balance classes
                    indices = np.random.choice(
                        indices, 
                        size=dataset_size,
                        replace=True,
                        p=sample_weights / np.sum(sample_weights)
                    )
                else:
                    # Fall back to simple shuffling if weighted sampling not available
                    np.random.shuffle(indices)
            
            # Yield batches
            for i in range(0, dataset_size, batch_size):
                batch_indices = indices[i:i + batch_size]
                
                # Skip last incomplete batch during training
                if len(batch_indices) < batch_size and is_train:
                    continue
                
                yield {
                    'input_ids': input_ids[batch_indices],
                    'attention_mask': attention_mask[batch_indices],
                    'labels': labels[batch_indices]
                }
        
        # Calculate number of batches
        num_batches = dataset_size // batch_size
        if dataset_size % batch_size > 0 and not is_train:
            num_batches += 1
        
        return dataloader, num_batches
    
    # Create dataloaders
    dataloaders = {}
    num_batches = {}
    
    for split in dataset:
        is_train = (split == 'train')
        dataloader, batches = create_dataloader(tokenized_datasets[split], is_train=is_train)
        dataloaders[split] = dataloader
        num_batches[split] = batches
    
    return dataloaders, num_batches





#----------------------------------------
# Improved Training Function with Gradient Accumulation
#----------------------------------------

def train_emotion_classifier(
    model,
    dataloaders,
    num_batches,
    num_epochs=15,
    learning_rate=5e-5,
    weight_decay=0.05,
    warmup_steps=200,
    decay_steps=None,
    seed=42,
    label_names=None,
    label_smoothing=0.1,
    gradient_accumulation_steps=2,
    use_ema=True,
    patience=8,
    visualize=True
):
    """
    Train a Mamba model for emotion classification with improved training dynamics.
    
    Args:
        model: The Mamba model to train
        dataloaders: Dictionary with train, validation, and test dataloaders
        num_batches: Dictionary with number of batches per dataloader
        num_epochs: Number of training epochs
        learning_rate: Learning rate
        weight_decay: Weight decay coefficient
        warmup_steps: Number of warmup steps
        decay_steps: Number of decay steps (defaults to training steps)
        seed: Random seed
        label_names: Names of emotion labels
        label_smoothing: Label smoothing factor
        gradient_accumulation_steps: Number of steps to accumulate gradients
        use_ema: Whether to use exponential moving average of parameters
        patience: Patience for early stopping
        visualize: Whether to generate visualizations
        
    Returns:
        results: Dictionary with trained model and metrics
    """
    # Set default label names if not provided
    if label_names is None:
        label_names = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
    
    # Calculate total training steps and decay steps if not provided
    total_train_steps = num_batches['train'] * num_epochs // gradient_accumulation_steps
    if decay_steps is None:
        decay_steps = total_train_steps - warmup_steps
    
    # Set random seed
    rng = jax.random.PRNGKey(seed)
    
    # Get input shape from first batch
    batch = next(dataloaders['train']())
    input_shape = batch['input_ids'].shape
    
    print(f"Input shape: {input_shape}")
    print(f"Number of labels: {model.num_labels}")
    print(f"Total training steps: {total_train_steps}")
    print(f"Warmup steps: {warmup_steps}")
    print(f"Decay steps: {decay_steps}")
    
    # Create train state
    state = create_train_state(
        model=model,
        rng=rng,
        learning_rate=learning_rate,
        weight_decay=weight_decay,
        input_shape=input_shape,
        warmup_steps=warmup_steps,
        decay_steps=decay_steps,
        label_smoothing=label_smoothing,
        use_ema=use_ema
    )
    
    # Initialize metrics history
    metrics_history = {
        'train_loss': [],
        'train_accuracy': [],
        'val_loss': [],
        'val_accuracy': [],
        'train_f1': [],
        'val_f1': [],
        'train_precision': [],
        'val_precision': [],
        'train_recall': [],
        'val_recall': []
    }
    
    # Initialize early stopping variables
    best_val_f1 = 0.0
    patience_counter = 0
    best_params = None
    
    # Print training info
    print(f"Training on {num_batches['train']} batches per epoch")
    print(f"Validating on {num_batches['validation']} batches per epoch")
    print(f"Testing on {num_batches['test']} batches per epoch")
    print(f"Using gradient accumulation: {gradient_accumulation_steps} steps")
    print(f"Label smoothing: {label_smoothing}")
    
    # Training loop
    for epoch in range(num_epochs):
        start_time = time.time()
        
        #---------- Training ----------#
        train_metrics = []
        train_labels = []
        train_preds = []
        
        # Create progress bar for training
        train_tqdm = tqdm(
            enumerate(dataloaders['train']()), 
            total=num_batches['train'],
            desc=f"Epoch {epoch+1}/{num_epochs} [Train]",
            leave=False
        )
        
        # Accumulation variables
        acc_step = 0
        accumulated_batches = []
        
        for batch_idx, batch in train_tqdm:
            # Convert batch to JAX arrays
            batch = {k: jnp.array(v) for k, v in batch.items()}
            
            # Accumulate batch
            accumulated_batches.append(batch)
            acc_step += 1
            
            # If enough batches accumulated or last batch, perform update
            if acc_step == gradient_accumulation_steps or batch_idx == num_batches['train'] - 1:
                # Process each batch
                for sub_batch in accumulated_batches:
                    # Perform training step
                    state, metrics = train_step(state, sub_batch, label_smoothing=label_smoothing)
                    train_metrics.append(metrics)
                    
                    # Collect labels and predictions for detailed metrics
                    train_labels.extend(sub_batch['labels'].tolist())
                    train_preds.extend(metrics['predictions'].tolist())
                
                # Reset accumulation variables
                acc_step = 0
                accumulated_batches = []
            
            # Calculate average metrics for display
            avg_loss = np.mean([m['loss'] for m in train_metrics]) if train_metrics else 0
            avg_acc = np.mean([m['accuracy'] for m in train_metrics]) if train_metrics else 0
            
            # Update progress bar
            train_tqdm.set_postfix({
                'loss': f"{avg_loss:.4f}", 
                'acc': f"{avg_acc:.4f}"
            })
        
        # Compute average training metrics
        train_metrics_avg = {
            k: float(np.mean([m[k] for m in train_metrics if k != 'predictions']))
            for k in train_metrics[0] if k != 'predictions'
        }
        
        # Compute detailed training metrics
        train_detailed_metrics = compute_detailed_metrics(
            true_labels=train_labels,
            predictions=train_preds,
            label_names=label_names
        )
        
        #---------- Validation ----------#
        val_metrics = []
        val_labels = []
        val_preds = []
        
        # Create progress bar for validation
        val_tqdm = tqdm(
            enumerate(dataloaders['validation']()), 
            total=num_batches['validation'],
            desc=f"Epoch {epoch+1}/{num_epochs} [Validation]",
            leave=False
        )
        
        for batch_idx, batch in val_tqdm:
            # Convert batch to JAX arrays
            batch = {k: jnp.array(v) for k, v in batch.items()}
            
            # Perform evaluation step
            metrics = eval_step(state, batch, use_ema=use_ema)
            val_metrics.append(metrics)
            
            # Collect labels and predictions for detailed metrics
            val_labels.extend(batch['labels'].tolist())
            val_preds.extend(metrics['predictions'].tolist())
            
            # Update progress bar
            val_tqdm.set_postfix({
                'loss': f"{metrics['loss']:.4f}", 
                'acc': f"{metrics['accuracy']:.4f}"
            })
        
        # Compute average validation metrics
        val_metrics_avg = {
            k: float(np.mean([m[k] for m in val_metrics if k != 'predictions']))
            for k in val_metrics[0] if k != 'predictions'
        }
        
        # Compute detailed validation metrics
        val_detailed_metrics = compute_detailed_metrics(
            true_labels=val_labels,
            predictions=val_preds,
            label_names=label_names
        )
        
        # Record metrics
        metrics_history['train_loss'].append(train_metrics_avg['loss'])
        metrics_history['train_accuracy'].append(train_metrics_avg['accuracy'])
        metrics_history['val_loss'].append(val_metrics_avg['loss'])
        metrics_history['val_accuracy'].append(val_metrics_avg['accuracy'])
        
        # Record F1, precision, and recall
        metrics_history['train_f1'].append(train_detailed_metrics['f1'])
        metrics_history['val_f1'].append(val_detailed_metrics['f1'])
        metrics_history['train_precision'].append(train_detailed_metrics['precision'])
        metrics_history['val_precision'].append(val_detailed_metrics['precision'])
        metrics_history['train_recall'].append(train_detailed_metrics['recall'])
        metrics_history['val_recall'].append(val_detailed_metrics['recall'])
        
        # Log metrics
        epoch_time = time.time() - start_time
        print(f"\nEpoch {epoch+1}/{num_epochs} | Time: {epoch_time:.2f}s")
        print(f"  Train Loss: {train_metrics_avg['loss']:.4f} | "
              f"Train Accuracy: {train_metrics_avg['accuracy']:.4f}")
        print(f"  Val Loss: {val_metrics_avg['loss']:.4f} | "
              f"Val Accuracy: {val_metrics_avg['accuracy']:.4f}")
        
        # Print detailed metrics
        print("\nTraining F1: {:.4f} | Precision: {:.4f} | Recall: {:.4f}".format(
            train_detailed_metrics['f1'],
            train_detailed_metrics['precision'],
            train_detailed_metrics['recall']
        ))
        print("Validation F1: {:.4f} | Precision: {:.4f} | Recall: {:.4f}".format(
            val_detailed_metrics['f1'],
            val_detailed_metrics['precision'],
            val_detailed_metrics['recall']
        ))
        
        # Early stopping check using F1 score (more comprehensive than accuracy)
        if val_detailed_metrics['f1'] > best_val_f1:
            best_val_f1 = val_detailed_metrics['f1']
            # Store the better parameters (EMA if available)
            if use_ema and state.ema_params is not None:
                best_params = jax.tree_util.tree_map(lambda x: x.copy(), state.ema_params)
            else:
                best_params = jax.tree_util.tree_map(lambda x: x.copy(), state.params)
            patience_counter = 0
            print(f"New best validation F1: {best_val_f1:.4f}")
        else:
            patience_counter += 1
            print(f"No improvement in validation F1. Patience: {patience_counter}/{patience}")
            
        if patience_counter >= patience:
            print(f"Early stopping triggered at epoch {epoch+1}")
            if best_params is not None:
                # Restore best params
                if use_ema and state.ema_params is not None:
                    state = state.replace(params=best_params, ema_params=best_params)
                else:
                    state = state.replace(params=best_params)
                print("Restored best model parameters")
            break
    
    # Use best parameters from early stopping
    if best_params is not None and patience_counter >= patience:
        print("Using best model from early stopping")
    
    #---------- Final Test Evaluation ----------#
    print("\nEvaluating on test set...")
    test_metrics = []
    test_labels = []
    test_preds = []
    
    # Create progress bar for testing
    test_tqdm = tqdm(
        enumerate(dataloaders['test']()), 
        total=num_batches['test'],
        desc="Test Evaluation"
    )
    
    for batch_idx, batch in test_tqdm:
        # Convert batch to JAX arrays
        batch = {k: jnp.array(v) for k, v in batch.items()}
        
        # Perform evaluation step with best parameters (no smoothing)
        metrics = eval_step(state, batch, use_ema=use_ema)
        test_metrics.append(metrics)
        
        # Collect labels and predictions for detailed metrics
        test_labels.extend(batch['labels'].tolist())
        test_preds.extend(metrics['predictions'].tolist())
        
        # Update progress bar
        test_tqdm.set_postfix({
            'loss': f"{metrics['loss']:.4f}", 
            'acc': f"{metrics['accuracy']:.4f}"
        })
    
    # Compute average test metrics
    test_metrics_avg = {
        k: float(np.mean([m[k] for m in test_metrics if k != 'predictions']))
        for k in test_metrics[0] if k != 'predictions'
    }
    
    # Compute detailed test metrics
    test_detailed_metrics = compute_detailed_metrics(
        true_labels=test_labels,
        predictions=test_preds,
        label_names=label_names
    )
    
    # Add test metrics to history
    metrics_history['test_loss'] = test_metrics_avg['loss']
    metrics_history['test_accuracy'] = test_metrics_avg['accuracy']
    metrics_history['test_f1'] = test_detailed_metrics['f1']
    metrics_history['test_precision'] = test_detailed_metrics['precision']
    metrics_history['test_recall'] = test_detailed_metrics['recall']
    
    # Print test results
    print(f"\nTest results:")
    print(f"  Loss: {test_metrics_avg['loss']:.4f} | "
          f"Accuracy: {test_metrics_avg['accuracy']:.4f}")
    print(f"  F1: {test_detailed_metrics['f1']:.4f} | "
          f"Precision: {test_detailed_metrics['precision']:.4f} | "
          f"Recall: {test_detailed_metrics['recall']:.4f}")
    
    # Print detailed test metrics
    print_detailed_metrics(test_detailed_metrics, title="Detailed Test Metrics")
    
    # Generate visualizations if requested
    if visualize:
        # Plot confusion matrix for test data
        plot_confusion_matrix(
            true_labels=test_labels,
            predictions=test_preds,
            label_names=label_names,
            save_path="test_confusion_matrix.png"
        )
        
        # Plot training history
        plot_training_history(
            metrics_history=metrics_history,
            save_path="training_history.png"
        )
    
    # Return results
    params_to_return = state.ema_params if use_ema and state.ema_params is not None else state.params
    return {
        'model': model,
        'params': params_to_return,
        'metrics_history': metrics_history,
        'test_metrics': test_metrics_avg,
        'test_detailed_metrics': test_detailed_metrics,
        'label_names': label_names
    }


#----------------------------------------
# Visualization and Evaluation Utilities with Improvements
#----------------------------------------

def compute_detailed_metrics(true_labels, predictions, label_names=None):
    """
    Compute detailed metrics including precision, recall, and F1 score.
    
    Args:
        true_labels: True labels
        predictions: Predicted labels
        label_names: Names of labels (optional)
        
    Returns:
        metrics: Dictionary with detailed metrics
    """
    # Compute overall metrics
    accuracy = np.mean(predictions == true_labels)
    
    # Compute F1, precision, and recall
    # Add zero_division=0 to handle undefined precision
    f1 = f1_score(true_labels, predictions, average='weighted', zero_division=0)
    precision = precision_score(true_labels, predictions, average='weighted', zero_division=0)
    recall = recall_score(true_labels, predictions, average='weighted', zero_division=0)
    
    # Compute per-class metrics
    # Add zero_division=0 to handle undefined precision
    class_f1 = f1_score(true_labels, predictions, average=None, zero_division=0)
    class_precision = precision_score(true_labels, predictions, average=None, zero_division=0)
    class_recall = recall_score(true_labels, predictions, average=None, zero_division=0)
    
    # Create per-class metrics dictionary
    per_class_metrics = {}
    for i, (f, p, r) in enumerate(zip(class_f1, class_precision, class_recall)):
        class_name = label_names[i] if label_names and i < len(label_names) else f"Class {i}"
        per_class_metrics[class_name] = {
            'f1': f,
            'precision': p,
            'recall': r
        }
    
    return {
        'accuracy': accuracy,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'per_class': per_class_metrics
    }


def print_detailed_metrics(metrics, title="Metrics"):
    """
    Print detailed metrics.
    
    Args:
        metrics: Dictionary with detailed metrics
        title: Title for the metrics section
    """
    print(f"\n{title}:")
    print(f"  Accuracy:  {metrics['accuracy']:.4f}")
    print(f"  F1 Score:  {metrics['f1']:.4f}")
    print(f"  Precision: {metrics['precision']:.4f}")
    print(f"  Recall:    {metrics['recall']:.4f}")
    
    print("\nPer-class metrics:")
    for class_name, class_metrics in metrics['per_class'].items():
        print(f"  {class_name}:")
        print(f"    F1:        {class_metrics['f1']:.4f}")
        print(f"    Precision: {class_metrics['precision']:.4f}")
        print(f"    Recall:    {class_metrics['recall']:.4f}")


def plot_confusion_matrix(true_labels, predictions, label_names=None, save_path=None):
    """
    Plot confusion matrix with improved visualization.
    
    Args:
        true_labels: True labels
        predictions: Predicted labels
        label_names: Names of labels (optional)
        save_path: Path to save the plot (optional)
    """
    # Compute confusion matrix
    cm = confusion_matrix(true_labels, predictions)
    
    # Normalize confusion matrix
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    # Set up plot with improved styling
    plt.figure(figsize=(12, 10))
    
    # Use a better colormap and improved annotations
    ax = sns.heatmap(
        cm_norm, 
        annot=cm,
        fmt='d',
        cmap='Blues',
        xticklabels=label_names if label_names else "auto",
        yticklabels=label_names if label_names else "auto",
        annot_kws={"size": 12, "weight": "bold"},
        linewidths=0.5,
        cbar_kws={"shrink": 0.8}
    )
    
    # Improve axis labels and title
    plt.xlabel('Predicted labels', fontsize=14, weight='bold')
    plt.ylabel('True labels', fontsize=14, weight='bold')
    plt.title('Confusion Matrix', fontsize=16, weight='bold')
    
    # Rotate tick labels for better readability
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    
    # Adjust layout
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        # plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Confusion matrix plot saved to {save_path}")
    
    plt.show()


def plot_training_history(metrics_history, save_path=None):
    """
    Plot training history with improved visualization.
    
    Args:
        metrics_history: Dictionary with training metrics history
        save_path: Path to save the plot (optional)
    """
    # Create a figure with three subplots
    plt.figure(figsize=(18, 8))
    
    # Plot loss
    plt.subplot(1, 3, 1)
    plt.plot(metrics_history['train_loss'], 'b-', linewidth=2, label='Train')
    plt.plot(metrics_history['val_loss'], 'r-', linewidth=2, label='Validation')
    if 'test_loss' in metrics_history:
        plt.axhline(y=metrics_history['test_loss'], color='g', linestyle='--', 
                   linewidth=2, label=f"Test ({metrics_history['test_loss']:.4f})")
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.legend(fontsize=10)
    plt.title('Loss', fontsize=14, weight='bold')
    plt.grid(True, alpha=0.3)
    
    # Plot accuracy
    plt.subplot(1, 3, 2)
    plt.plot(metrics_history['train_accuracy'], 'b-', linewidth=2, label='Train')
    plt.plot(metrics_history['val_accuracy'], 'r-', linewidth=2, label='Validation')
    if 'test_accuracy' in metrics_history:
        plt.axhline(y=metrics_history['test_accuracy'], color='g', linestyle='--', 
                   linewidth=2, label=f"Test ({metrics_history['test_accuracy']:.4f})")
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.legend(fontsize=10)
    plt.title('Accuracy', fontsize=14, weight='bold')
    plt.grid(True, alpha=0.3)
    
    # Plot F1 score
    plt.subplot(1, 3, 3)
    plt.plot(metrics_history['train_f1'], 'b-', linewidth=2, label='Train')
    plt.plot(metrics_history['val_f1'], 'r-', linewidth=2, label='Validation')
    if 'test_f1' in metrics_history:
        plt.axhline(y=metrics_history['test_f1'], color='g', linestyle='--', 
                   linewidth=2, label=f"Test ({metrics_history['test_f1']:.4f})")
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('F1 Score', fontsize=12)
    plt.legend(fontsize=10)
    plt.title('F1 Score', fontsize=14, weight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        # plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Training history plot saved to {save_path}")
    
    # Additional detailed metrics plot (precision/recall)
    plt.figure(figsize=(15, 6))
    
    # Plot precision
    plt.subplot(1, 2, 1)
    plt.plot(metrics_history['train_precision'], 'b-', linewidth=2, label='Train')
    plt.plot(metrics_history['val_precision'], 'r-', linewidth=2, label='Validation')
    if 'test_precision' in metrics_history:
        plt.axhline(y=metrics_history['test_precision'], color='g', linestyle='--', 
                   linewidth=2, label=f"Test ({metrics_history['test_precision']:.4f})")
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.legend(fontsize=10)
    plt.title('Precision', fontsize=14, weight='bold')
    plt.grid(True, alpha=0.3)
    
    # Plot recall
    plt.subplot(1, 2, 2)
    plt.plot(metrics_history['train_recall'], 'b-', linewidth=2, label='Train')
    plt.plot(metrics_history['val_recall'], 'r-', linewidth=2, label='Validation')
    if 'test_recall' in metrics_history:
        plt.axhline(y=metrics_history['test_recall'], color='g', linestyle='--', 
                   linewidth=2, label=f"Test ({metrics_history['test_recall']:.4f})")
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Recall', fontsize=12)
    plt.legend(fontsize=10)
    plt.title('Recall', fontsize=14, weight='bold')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        precision_recall_path = save_path.replace('.png', '_precision_recall.png')
        # plt.savefig(precision_recall_path, dpi=300, bbox_inches='tight')
        print(f"Precision/Recall plot saved to {precision_recall_path}")
    
    plt.show()


#----------------------------------------
# Helper Functions for Using the Model with Improvements
#----------------------------------------

def save_model_params(params, filename="mamba_emotion_params.pkl", include_metadata=True, metadata=None):
    """
    Save model parameters to a file with additional metadata.
    
    Args:
        params: Model parameters
        filename: Output filename
        include_metadata: Whether to include metadata
        metadata: Additional metadata to include
    """
    if include_metadata:
        # Create a dictionary with parameters and metadata
        save_dict = {
            'params': params,
            'metadata': metadata or {},
            'saved_at': time.strftime('%Y-%m-%d %H:%M:%S')
        }
        with open(filename, "wb") as f:
            pickle.dump(save_dict, f)
    else:
        # Save only parameters
        with open(filename, "wb") as f:
            pickle.dump(params, f)
    
    print(f"Model parameters saved to {filename}")


def load_model_params(filename="mamba_emotion_params.pkl", return_metadata=False):
    """
    Load model parameters from a file.
    
    Args:
        filename: Input filename
        return_metadata: Whether to return metadata
        
    Returns:
        params: Model parameters
        metadata: Metadata (if return_metadata=True)
    """
    with open(filename, "rb") as f:
        loaded = pickle.load(f)
    
    # Check if it's a dictionary with metadata or just parameters
    if isinstance(loaded, dict) and 'params' in loaded:
        params = loaded['params']
        metadata = loaded.get('metadata', {})
        print(f"Model parameters loaded from {filename} (saved at {loaded.get('saved_at', 'unknown time')})")
        
        if return_metadata:
            return params, metadata
        return params
    else:
        # Legacy format - just parameters
        print(f"Model parameters loaded from {filename} (legacy format)")
        if return_metadata:
            return loaded, {}
        return loaded


def predict_emotion(text, model, params, tokenizer, label_names=None):
    """
    Predict emotion for a given text with confidence scores.
    
    Args:
        text: Input text
        model: Trained model
        params: Model parameters
        tokenizer: Tokenizer
        label_names: Names of emotion labels
        
    Returns:
        prediction: Predicted emotion with confidence scores
    """
    # Default label names
    if label_names is None:
        label_names = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
    
    # Clean and standardize text
    text = text.strip().replace('\n', ' ').replace('\r', ' ')
    
    # Tokenize text
    tokens = tokenizer(
        text,
        max_length=96,  # Reduced from 128
        padding='max_length',
        truncation=True,
        return_tensors='np'
    )
    
    # Convert to JAX arrays
    input_ids = jnp.array(tokens['input_ids'])
    attention_mask = jnp.array(tokens['attention_mask'])
    
    # Make prediction
    logits = model.apply(
        {'params': params},
        input_ids,
        attention_mask,
        deterministic=True
    )
    
    # Get predicted class
    pred_class = jnp.argmax(logits, axis=-1)[0]
    
    # Convert to emotion name
    emotion = label_names[pred_class]
    
    # Get probabilities
    probs = jax.nn.softmax(logits, axis=-1)[0]
    prob = float(probs[pred_class])
    
    # Get all emotion probabilities
    all_emotions = {label_names[i]: float(probs[i]) for i in range(len(label_names))}
    
    # Get top 3 emotions
    top_emotions = sorted(all_emotions.items(), key=lambda x: x[1], reverse=True)[:3]
    
    return {
        'emotion': emotion,
        'probability': prob,
        'confidence': 'high' if prob > 0.7 else ('medium' if prob > 0.4 else 'low'),
        'top_emotions': top_emotions,
        'emotions': all_emotions
    }


#----------------------------------------
# Complete Function to Run the Improved Model
#----------------------------------------

def train_and_evaluate_mamba_emotion_model(
    num_epochs=15, 
    batch_size=32, 
    d_model=96,  # Reduced from 128
    n_layer=3,   # Reduced from 4
    dropout_rate=0.3,  # Increased from 0.1/0.2
    learning_rate=5e-5,  # Further reduced
    weight_decay=0.05,  # Increased from 0.01
    label_smoothing=0.1,  # Add label smoothing
    save_model=True,
    use_ema=True
):
    """
    Complete function to train and evaluate the improved Mamba model on the emotion dataset.
    
    Args:
        num_epochs: Number of training epochs
        batch_size: Batch size
        d_model: Model dimension
        n_layer: Number of layers
        dropout_rate: Dropout rate
        learning_rate: Learning rate
        weight_decay: Weight decay coefficient
        label_smoothing: Label smoothing factor
        save_model: Whether to save the model
        use_ema: Whether to use exponential moving average
        
    Returns:
        results: Dictionary with trained model and metrics
    """
    # Set up device info
    device_info = setup_jax_devices()
    
    # Import tokenizer
    tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
    
    # Define emotion label names
    label_names = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
    
    # Load dataset
    print("Loading emotion dataset...")
    dataset = load_emotion_dataset()
    
    # Prepare dataset for training with augmentation
    print("Preparing dataset for training...")
    dataloaders, num_batches = prepare_dataset_for_training(
        dataset=dataset,
        tokenizer=tokenizer,
        max_length=96,  # Reduced from 128
        batch_size=batch_size,
        augment=True  # Enable augmentation
    )
    
    # Create improved model
    print("Creating improved Mamba model...")
    model = MambaForSequenceClassification(
        vocab_size=tokenizer.vocab_size,
        num_labels=len(label_names), 
        d_model=d_model,
        n_layer=n_layer,
        d_state=12,  
        expand=1,    
        dropout_rate=dropout_rate,
        classifier_dropout=dropout_rate * 1.5,
        embedding_dropout=0.1, 
        label_smoothing=label_smoothing
    )
    
    # Calculate warmup and decay steps
    total_steps = num_batches['train'] * num_epochs
    warmup_steps = int(0.1 * total_steps)  
    decay_steps = total_steps - warmup_steps
    
    # Train and evaluate model
    print("Training and evaluating model...")
    results = train_emotion_classifier(
        model=model,
        dataloaders=dataloaders,
        num_batches=num_batches,
        num_epochs=num_epochs,
        learning_rate=learning_rate,
        weight_decay=weight_decay,
        warmup_steps=warmup_steps,
        decay_steps=decay_steps,
        label_smoothing=label_smoothing,
        gradient_accumulation_steps=2, 
        use_ema=use_ema,  
        label_names=label_names
    )
    
    # Save the model
    if save_model:
        # Include metadata
        metadata = {
            'model_config': {
                'd_model': d_model,
                'n_layer': n_layer,
                'dropout_rate': dropout_rate,
                'label_smoothing': label_smoothing
            },
            'training_config': {
                'num_epochs': num_epochs,
                'batch_size': batch_size,
                'learning_rate': learning_rate,
                'weight_decay': weight_decay,
                'warmup_steps': warmup_steps,
                'decay_steps': decay_steps,
                'use_ema': use_ema
            },
            'performance': {
                'test_accuracy': results['test_metrics']['accuracy'],
                'test_f1': results['test_detailed_metrics']['f1'],
                'test_precision': results['test_detailed_metrics']['precision'],
                'test_recall': results['test_detailed_metrics']['recall']
            },
            'label_names': label_names
        }
        
        save_model_params(
            results['params'], 
            "improved_mamba_emotion_model.pkl",
            include_metadata=True,
            metadata=metadata
        )
        print("Model saved to improved_mamba_emotion_model.pkl")
    
    # Test with some example texts
    print("\nTesting with example texts:")
    examples = [
        "I am so happy today!",
        "I'm feeling quite sad and sorry for myself",
        "I can't believe how angry I am about this situation",
        "I'm so scared of what might happen next",
        "Wow! That's such a surprise!",
        "I love spending time with my family"
    ]
    
    for text in examples:
        # Make prediction
        result = predict_emotion(
            text=text,
            model=model,
            params=results['params'],
            tokenizer=tokenizer,
            label_names=label_names
        )
        
        print(f"\nText: \"{text}\"")
        print(f"Predicted emotion: {result['emotion']} (confidence: {result['probability']:.2f} - {result['confidence']})")
        
        # Print top 3 emotions
        print("Top emotions:")
        for emotion, prob in result['top_emotions']:
            print(f"  {emotion}: {prob:.4f}")
    
    return results


if __name__ == "__main__":
    train_and_evaluate_mamba_emotion_model(
        num_epochs=25,          
        batch_size=32,          
        d_model=96,             
        n_layer=3,              
        dropout_rate=0.3,       
        learning_rate=5e-5,     
        weight_decay=0.05,
        label_smoothing=0.1,  
        save_model=True,
        use_ema=True
    )
```
